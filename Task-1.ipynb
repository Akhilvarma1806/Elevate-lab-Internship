```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\pc\Desktop\Elevate Labs\archive\patients.csv")
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.drop_duplicates()
```

```python
df['age']=df['age'].astype(str).str.extract('(\d+)')
df['age']=pd.to_numeric(df['age'],errors='coerce')
df['age'].fillna(df['age'].median(),inplace=True)
```

```python
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
df['departure_date'] = pd.to_datetime(df['departure_date'])
df['arrival_date'] = df['arrival_date'].dt.strftime('%d-%m-%Y')
df['departure_date'] = df['departure_date'].dt.strftime('%d-%m-%Y')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns = (
    df.columns
    .str.strip()                # remove leading/trailing spaces
    .str.lower()                # convert to lowercase
    .str.replace(' ', '_')      # replace spaces with underscores
    .str.replace('[^a-z0-9_]', '', regex=True)  # remove special chars
)
```
